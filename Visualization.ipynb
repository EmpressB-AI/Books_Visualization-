```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('bestsellers with categories.csv')
data.head(10)
```

# Data Visualizations

## The Best Selling Books of The Decade
**Using User Ratings and the names of books**

```python
data[['Name', 'User Rating']].nlargest(12, 'User Rating').plot.bar(x = 'Name', y = 'User Rating', color ='black')
plt.title('The best books of the Year', size=20)
plt.xlabel('Name of the books', fontsize=25, color='blue')
plt.ylabel('User Ratings', fontsize=25, color='blue')
```

## Percentage of Fiction to Non-fiction

```python
data['Genre'].value_counts()
```

```python
len(data[data['Genre'] == 'Fiction'])
```

```python
Data = [['Non Fiction', len(data[data['Genre'] == 'Non Fiction'])], 
        ['Fiction', len(data[data['Genre'] == 'Fiction'])]]
```

```python
slices = [len(data[data['Genre'] == 'Non Fiction']), len(data[data['Genre'] == 'Fiction'])]
print(slices)
labels = ['Non Fiction', 'Fiction']
colors = ['r', 'k']
plt.pie(slices, labels=labels, colors=colors, startangle=180)
```

## Top Ten Authors According To Reviews

```python
data[['Author', 'Reviews']].nlargest(12, 'Reviews').plot.bar(x = 'Author', y = 'Reviews', color ='black')
plt.title('Top Ten Authors According To Reviews', size=20)
plt.xlabel('Name of the Authors', fontsize=25, color='blue')
plt.ylabel('User Reviews', fontsize=25, color='blue')
```
